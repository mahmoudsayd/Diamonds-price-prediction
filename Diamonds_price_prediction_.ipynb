```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['clarity'] = le.fit_transform(df['clarity'])
df['color'] = le.fit_transform(df['cut'])
df['cut'] = le.fit_transform(df['cut'])

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler
```

```python
Y = df['price']
X = df.drop('price', axis=1)
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state =0)
```

```python
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)

x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
le = LinearRegression()
le.fit(x_train, y_train)
score1 = le.score(x_train, y_train)
print("train score : ", score1)
```

```python
y_pred1 = le.predict(x_test)
r_score1 = r2_score(y_test, y_pred1)
print("Testing Score : ", r_score1)
```

```python
from sklearn.linear_model import Lasso
lasso_regression = Lasso(alpha=0.8, max_iter=10000)
lasso_regression.fit(x_train, y_train)
score2 = lasso_regression.score(x_train, y_train)
print("train score : ", score2)
```

```python
y_pred2 = lasso_regression.predict(x_test)
r_score2 = r2_score(y_test, y_pred2)
print("Testing Score : ", r_score2)
```
